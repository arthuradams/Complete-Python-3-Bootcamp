```python
# Open the file

data = open("example.csv",encoding="utf-8")

# csv.reader

csv_data = csv.reader(data)

# reformat into a python list of lists

data_lines = list(csv_data)
```

```python
keys = data_lines.pop(0)
```

```python
keys
```

```python
dict = []
for line in data_lines:
    curr_line = {}
    curr_cell = 0
    for key in keys:
        curr_line[key] = line[curr_cell]
        curr_cell += 1
    dict.append(curr_line)
        
```
